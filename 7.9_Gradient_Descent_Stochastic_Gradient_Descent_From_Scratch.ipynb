```python
class StochasticGradientDescentRegressor:

    def __init__(self, learning_rate, epochs):
        self.learning_rate=learning_rate
        self.epochs=epochs
        self.coef_=None
        self.intercept_ = None

    def fit(self, X, y):
        self.intercept = 0
        self.coefficient = np.ones(X.shape[1])

        for i in range(self.epochs):
           for j in range(X.shape[0]):
               random_index = np.random.randint(0, X.shape[0])
               y_hat= np.dot(X[random_index], self.coefficient) + self.intercept
               deri_intercept = -2 * (y[random_index] - y_hat ) 
               self.intercept = self.intercept - (self.learning_rate * deri_intercept)
               deri_coefficient = - 2 * np.dot((y[random_index] - y_hat), X[random_index])
               self.coefficient = self.coefficient - (self.learning_rate * deri_coefficient )

                
        print(self.intercept, self.coefficient)

    def predict(sefl, X_test):
        return np.dot(X, self.coefficient) + self.intercept
        
```
