```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df['number'].value_counts().plot(kind='bar')
```

```python
df['number'].unique()
```

```python
df['numerical_data_number']=pd.to_numeric(df['number'], errors='coerce', downcast='integer')
```

```python
# Extracting Categorical Data
df['categorical_data_number']=np.where(df['numerical_data_number'].isnull(),df['number'], np.nan)
```

```python
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Cabin_letter']=df['Cabin'].str[0]
```

```python
df['Cabin_num']=df['Cabin'].str.extract('(\d+)')
```
