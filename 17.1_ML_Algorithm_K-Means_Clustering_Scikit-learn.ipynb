```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('Datasets/student_clustering.csv')
```

```python
df.sample(5)
```

```python
X = df.iloc[:, 0:1]
y = df['iq']
```

```python
plt.scatter( X, y, c=y, edgecolor = 'k')
```

```python
wcss = []
for i in range(1, 15) : 
    km = KMeans(n_clusters= i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
print(plt.plot(range(1, 15), wcss))
```
