```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
```

```python
X, y = make_classification(n_samples= 2000, n_features= 10, n_informative=3)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 40)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.tree import plot_tree

dt = DecisionTreeClassifier(max_depth=13)
dt.fit(X_train, y_train)
```

```python
y_predict = dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
a_s = accuracy_score(y_test, y_predict)
```

```python
print(f"Accuracy_score using decision tree : {a_s}")
```

```python
# plot_tree(dt)
```

```python
bc = BaggingClassifier(estimator = dt, n_estimators= 250, max_samples=0.2, bootstrap= True, random_state= 40)
```

```python
bc.fit(X_train, y_train)
```

```python
y_predict_2 = bc.predict(X_test)
```

```python
print(f"Accuracy_score using bagging , estimator as Decision Tree Classifier : {accuracy_score(y_test, y_predict_2)}")
```

```python
bc2 = BaggingClassifier(estimator = SVC(),
                        n_estimators = 250, 
                        max_samples=0.2 , 
                        bootstrap=True, 
                        random_state=40)
```

```python
bc2.fit(X_train, y_train)
```

```python
y_predict_3 = bc2.predict(X_test)
```

```python
print(f"Accuracy score using bagging , estimator as SVC : {accuracy_score(y_test, y_predict_3)}")
```
