```python
import numpy as np

import matplotlib.pyplot as plt

from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
```

```python
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=20)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='bwr',edgecolors = 'k')
plt.xlabel('feature 1')
plt.ylabel('feature 2')
plt.title('Binary Classification')
plt.show()
```

```python
def StepFunction(z):
    if z > 0:
        return 1 
    else:
        return 0
```

```python
def LogisticRegressionPerceptron(X, y):
    X = np.insert(X, 0, 1, axis = 1)
    weights = np.ones(X.shape[1])
    learning_rate = 0.1
    epochs = 1000

    for i in range(epochs):
        random_idx = np.random.randint(0, X.shape[0])
        y_predict = StepFunction(np.dot(X[random_idx], weights))
        weights = weights + learning_rate * (y[random_idx] - y_predict) * X[random_idx]
        
    return weights[0], weights[1:]
```

```python
intercept, coefficient = LogisticRegressionPerceptron(X, y)
```

```python
intercept
```

```python
coefficient
```

```python
m = -(coefficient[0]/coefficient[1])
c = -(intercept/coefficient[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + c
```

```python
plt.figure(figsize= (10, 6))
plt.plot(x_input, y_input, color = 'red', linewidth = 3)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap = 'bwr', edgecolor='k')
plt.ylim(-3, 2)
plt.xlabel('feature 1')
plt.ylabel('feature 2')
plt.title('Binary Classification')
plt.show()
```

```python
lor = LogisticRegression()
lor.fit(X,y)
```

```python
m = -(lor.coef_[0][0]/lor.coef_[0][1])
c = -(lor.intercept_/lor.coef_[0][1])
```

```python
x_input1 = np.linspace(-3,3,100)
y_input1 = m*x_input + c
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3, label = 'Step Function')
plt.plot(x_input1,y_input1,color='black',linewidth=3, label = 'Logistic Regression')
plt.scatter(X[:,0],X[:,1],c=y,cmap='bwr', edgecolor= 'k')
plt.ylim(-3,2)
plt.legend()
plt.title('Binary Classification')
plt.xlabel('feature 1')
plt.ylabel('feature 2')
plt.show()
```

```python
def SigmoidFunction(z):
    return 1/(1 + np.exp(-z))
```

```python
def LogisticRegressionSigmoid(X,y):
    
    X = np.insert(X, 0, 1, axis = 1)
    weights = np.ones(X.shape[1])
    learning_rate = 0.1
    epochs = 1000

    for i in range(epochs):
        random_idx = np.random.randint(0, X.shape[0])
        y_predict = SigmoidFunction(np.dot(X[random_idx], weights))
        weights = weights + learning_rate * (y[random_idx] - y_predict) * X[random_idx]
        
    return weights[0], weights[1:]
```

```python
intercept_,coef_ = LogisticRegressionSigmoid(X,y)
```

```python
m = -(coef_[0]/coef_[1])
c = -(intercept_/coef_[1])
```

```python
x_input2 = np.linspace(-3,3,100)
y_input2 = m*x_input + c
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3, label = 'Step Function')
plt.plot(x_input1,y_input1,color='black',linewidth=3, label = 'Sigmoid Function')
plt.plot(x_input2,y_input2,color='brown',linewidth=3, label = ' Logistic Regression')
plt.scatter(X[:,0],X[:,1],c=y,cmap='bwr', edgecolor = 'k')
plt.ylim(-3,2)
plt.legend()
plt.xlabel('feature 1')
plt.ylabel('feature 2')
plt.title('Binary Classification')
```
